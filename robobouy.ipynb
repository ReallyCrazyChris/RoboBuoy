```python
def relativeHeading(heading_from, heading_to):
    ''' Yields the angle difference in degrees between two headings'''

    relative_heading = heading_to - heading_from

    # constrain to -180 to 180 degree range
    if (relative_heading > 180):
        relative_heading -= 360
    
    if (relative_heading < -180):
        relative_heading += 360

    return relative_heading

def constrain(heading):
    '''heading is constrained to -180 to 180 degree range'''
    if (heading > 180):
        heading -= 360
    
    if (heading < -180):
        heading += 360
    return heading    



compassalpha = 0.97

def fusecompass(compasscourse, currentcourse):
    '''fuses the compass course witht he current course using a complement filter, strongly weighted towards the gyro'''
    currentcourse = compassalpha * currentcourse - (1.0 - compassalpha) * compasscourse 
    # clamp to -180 ... 180 degrees
    currentcourse = constrain(currentcourse)    
    return currentcourse
```

```python
''' Builds basic Getters and setters from stores keys'''

from lib.store import Store
store = Store()
keys = store.__dict__.keys()

for key in keys:
    print('@property',sep='')   
    print('def ',key,'(self):',sep='')
    print('\treturn self._',key,sep='') 
    print('')
    print('@',key,'.setter',sep='')   
    print('def ',key,'(self, value):',sep='')
    print('\tself._',key,' = value',sep='')
    print('')
    print('')   
```

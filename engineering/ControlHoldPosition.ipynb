## Requirement
- The robot approaches the target, which is shown by the target distance
- Once the robot enters the approachRadius, the robots velocity is reduces
- The Robot velocity will be 0 when it reaches the target
- It is not expected that the robot will actually reach the target.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define targetDistance
targetDistance = 10  # You can change this value as needed

# Calculate surge values
i_values = np.arange(0,10,0.1)
surge_values = [ min(1,(i**2)*0.02) for i in i_values]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(i_values, surge_values, 'b-', linewidth=2, marker='o', markersize=5)
#plt.grid(True, linestyle='--', alpha=0.7)

# Add labels and title
plt.xlabel('i value', fontsize=12)
plt.ylabel('Surge Value', fontsize=12)
plt.title(f'Surge Value vs. i (targetDistance = {targetDistance})', fontsize=14)

# Add text annotation explaining the formula
#plt.figtext(0.5, 0.01, f'Formula: surge = (targetDistance*100 - i) / 100', 
#           ha='center', fontsize=10, bbox={'facecolor':'lightgray', 'alpha':0.5, 'pad':5})

# Show the plot
#plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()
```


## Simplificaiton
From the graph above we simplify the code to:

```python
targetDistance = 20
targetRadius  = 5

for distance in range(targetDistance,0,-1):
        surge = min(1,(distance/targetRadius)**2)
        print(f"Hold: Distance: {distance:.2f}m, surge={surge:.2f}, ")
     
```

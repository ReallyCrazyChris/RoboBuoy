```python
    
class PID():
    def __init__(self, kP=1.0, kI=0.01, kD=0.01, dT=0.1 ):

        self.setpoint = 50

        self.Kp = 0.1 #kP
        self.Ki = 0.05 #kI
        self.Kd = 0.001 #kD
        self.dt = 1 #dT  # seconds

        self.error = 0
        self.errSum = 0
        self.dErr = 0      
        self.lastErr = 0  # Previous error  

    def loop(self, input):

        self.error  = self.setpoint - input
        self.errSum = self.errSum + (self.error * self.dt)
        self.dErr = (self.error - self.lastErr) / self.dt
        
        output = (self.Kp * self.error) + (self.Ki * self.errSum) + (self.Kd * self.dErr)

        self.lastErr = self.error

        return output
```

```python
pid = PID()
input = 100
for i in range(1,1000):

    output = pid.loop( input )

    input = input + output

    print( input ,output )
```
